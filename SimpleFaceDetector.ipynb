# Simple Face Detector

The following is a program that takes a live image of a face and draws a box around certain coordinates of the image. 

For the first week deliverable this is a hardcoded version, later on I will implement it using HAAR cascade.

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Note: this code has been taken from: https://github.com/googlecolab/colabtools/issues/560
# Accessing the camera using Colab is difficult because it uses a VM, thus we need to access it through Web API

from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def snap_face(filename='face.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('face/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

```python
def draw_box(path, upper_left, bottom_right, color, thickness):

  img = cv2.imread(path)

  img = cv2.resize(img, (img.shape[0], img.shape[0]))

  print('img.shape = ', img.shape)

  img = cv2.rectangle(img, upper_left, bottom_right, color, thickness)

  cv2_imshow(img)
```

```python
# define hyperparameters

path = 'face.jpg'

upper_left = (110, 110)
bottom_right = (380, 380)

color = (0, 0, 255)
thickness = 2

# snap_face()
draw_box(path, upper_left, bottom_right, color, thickness)
```
